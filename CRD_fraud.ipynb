__*The Dataset is highly unbalanced*__

0 = Normal Transaction;

__*Separate the data*__

```python
normal = df[df.Class == 0]
fraud = df[df.Class == 1]
```

__*Balaning the Dataset using Under-Sampling method*__

```python
#Number of Fraudulent Transactions --> 473
normal_sample = normal.sample(n=473)
```

```python
new_df = pd.concat([normal_sample, fraud], axis=0)
```

```python
new_df
```

__*The dataset is balanced*__

```python
new_df.groupby('Class').mean()
```

__*Splitting the data into Features & Targets*__

```python
X = new_df.drop(columns='Class', axis=1)
Y = new_df['Class']
```

```python
X
```

```python
Y
```

__*Split the data into Training data & Testing Data*__

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
X.shape, X_train.shape, X_test.shape
```

__*Model Training using Logistic Regression*__

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```
